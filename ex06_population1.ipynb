### 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt
with open("./datas/age_2023.csv",'r') as f:
    data = csv.reader(f)
    next(data)
    result=[] # 인구 저장할 리스트

    for row in data:
        if '이문제1동' in row[0]:
            for i in row[3:]: # 3번 인덱스 ~ 끝까지 반복
                result.append(int(i.replace(",",'')))
print(len(row[3:]))
# print(result)

plt.style.use("ggplot")
plt.rc('font',family='Malgun Gothic') # 한글 글꼴 추가
plt.plot(result,'r') 
plt.xlabel("연령")
plt.ylabel("인구")
plt.show()
```
