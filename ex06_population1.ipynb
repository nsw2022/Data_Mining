### 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt
with open("./datas/age_2023.csv",'r') as f:
    data = csv.reader(f)
    next(data)
    result=[] # 인구 저장할 리스트
    name = input("지역의 이름(읍면동 단위)을 입력해주세요 ")
    
    for row in data:
        if name in row[0]:
            for i in row[3:]: # 3번 인덱스 ~ 끝까지 반복
                result.append(int(i.replace(",",'')))
print('전체데이터수',len(row[3:]))
# print(result)

plt.style.use("ggplot")
plt.title(name + " 지역의 인구 구조")
plt.rc('font',family='Malgun Gothic') # 한글 글꼴 추가
# plt.plot(result,'r') 
plt.bar(range(0, 101), result) # x축 -range(0,101) , y축 - result
plt.xlabel("연령")
plt.ylabel("인구")
plt.show()
```
