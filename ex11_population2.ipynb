## 우리 동네 지역의 남여 성별 인구 분포

```python
import csv
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

with open("./datas/gender.csv") as f:
    data=csv.reader(f)
    header = next(data)
    
    m=[]#연령별 남성인구
    fm = []#연령별 여성인구
    
    for row in data:
#         print(row)
        if '대치1동' in row[0]:
            print("남성총인구",row[2],"여성총인구",row[105]) # 남성 총인구, 여성 총인구
            for i in row[3:104]:  # 0~100(101개)
                m.append(int(i)) # 숫자로 변환해서 저장
            for i in row[106:]:
                fm.append(int(i))
            break
# print(m)
# print(fm)
# 그래프로 출력
plt.plot(m,'red')
plt.plot(fm,'blue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
#Malgun Gothic
plt.rc('font', family='Batang') # 글꼴깨짐으로인한 폰트추가
plt.rcParams['axes.unicode_minus'] = False # 차트 -를 잡아주기 위한 설정
with open("./datas/gender.csv") as f:
    data=csv.reader(f)
    header = next(data)
    
    m=[]#연령별 남성인구
    fm = []#연령별 여성인구
    name = input("찾고 싶은 지역명을 입력하세요: ") # <- 지역명 입력받음
    
    for row in data:
#         print(row)
        if name in row[0]:
            print("남성총인구",row[2],"여성총인구",row[105]) # 남성 총인구, 여성 총인구
            for i in row[3:104]:  # 0~100(101개)
                # 숫자로 변환해서 저장 및 천단위 구분쉼표 제거
                m.append(-int(i.replace(",",""))) 
            for i in row[106:]:
                fm.append(int(i.replace(",","")))
            break
# print(m)
# print(fm)
# 그래프로 출력
plt.title(name + " 지역의 남여 성별 인구 분포")
plt.barh(range(101), m, label="남성") # 가로 막대형 그래프
plt.barh(range(101), fm, label="여성")
plt.legend() # <- 이게 있어야 라벨 들어감
plt.show()
```
